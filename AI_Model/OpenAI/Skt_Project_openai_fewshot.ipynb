```python
import csv
import openai
import numpy as np

# Set your OpenAI API key
openai.api_key = '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'

# Function to generate text based on the facts
def generate_text_based_on_facts(facts, max_tokens=100):
    prompt = f"{facts} 사례에서, 민사면 0 형사면 1을 리턴해줘, 숫자만 출력."
    response = openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",
        prompt=prompt,
        max_tokens=max_tokens
    )
    return response.choices[0].text.strip()

# Function to process the CSV file and calculate accuracy
def process_csv_and_calculate_accuracy(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        data = list(reader)

    # Selecting the first 50 and last 50 questions
    questions = [row[0] for row in data[:50] + data[-50:]]
    true_labels = [int(row[2]) for row in data[:50] + data[-50:]]  # Assuming that labels are in column 'C'

    # Generate predictions using the model
    predictions = []
    for fact in questions:
        generated_text = generate_text_based_on_facts(fact)
        # Assuming that '민사' in the response denotes a civil case (label 0) and '형사' denotes a criminal case (label 1)
        prediction = 0 if '민사' in generated_text else 1
        predictions.append(prediction)

    # Calculating accuracy
    accuracy = np.mean(np.array(predictions) == np.array(true_labels))
    return accuracy

# Example usage
file_path = '/content/reduced_lawsuit_type.csv'
accuracy = process_csv_and_calculate_accuracy(file_path)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
